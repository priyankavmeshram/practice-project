we can see the otliers in the columns bmi and charges.
but as charges is our target column, we cannot do anything to it.
we will try to remove the outliers from bmi column.

so now removing outlier using zscore method

```python
from scipy.stats import stats

z= np.abs(stats.zscore(df['bmi']))
print(z)
```

```python
threshold_z = 2
outlier = np.where(z > threshold_z)[0]
outlier
```

```python
df1 = df.drop(outlier)
df1.shape
```

```python
print("Data loss percentage- ",((df.shape[0]-df1.shape[0])/df.shape[0])*100)
```

```python
df1
```

now using IQR method

```python
#now using IQR method

#1st quantile
Q1 = df['bmi'].quantile(0.25)

#3rd quantile
Q3 = df['bmi'].quantile(0.75)

#IQR
IQR = Q3 - Q1

df2= df[~((df <(Q1 - 1.5 * IQR))|(df> (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
df2.shape
```

```python
print("Data loss percentage- ",((df.shape[0]-df2.shape[0])/df.shape[0])*100)
```

# seprating Feature and Label Column


```python
#seprating feature and label into x and y variable
x = df.drop("charges", axis=1)
y = df["charges"]

print("Feature Dimension =", x.shape)
print("Target Dimension =", y.shape)
```

# Feature scaling using Standard Scalarization

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
x=pd.DataFrame(scaler.fit_transform(x), columns = x.columns)
x
```

We have scaled the data using standard scalarization to overcome the issue of biasness.

# Checking Variance Inflation Factor(VIF)


```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif= pd.DataFrame()
vif["VIF values"]= [variance_inflation_factor(x.values, i) for  i in range(len(x.columns))]
vif["Features"]= x.columns
vif
```

From the above we can see that there are no columns who are more than the threshold value, so it means that there is no multicolinerity problem.

# Modelling

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
```

Finding the best random state

```python
# creating train tes split
maxAccu=0
maxRs=0
for i in range(1,200):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=40, random_state=i)
    lr= LinearRegression()
    lr.fit(x_train, y_train)
    pred = lr.predict(x_test)
    acc = r2_score(y_test, pred)
    if acc>maxAccu:
        maxAccu=acc
        maxRs=i
print("Maximum r2 score is", maxAccu,"On Random State", maxRs)
    
```
