#  checking for the outliers

```python
# cheking the outliers with the help of boxplot

plt.figure(figsize= (20,15), facecolor='white')
plotnumber = 1
for column in df:
    if plotnumber<=12:
        ax = plt.subplot(3,4, plotnumber)
        sns.boxplot(df[column], palette = 'Set2_r')
        plt.xlabel(column, fontsize=20)
        plt.yticks(rotation = 0, fontsize = 20)
    plotnumber+=1
plt.tight_layout()
plt.show()
```

As we can see that the outliers in all the columns,so now we can remove these outliers

```python
#divide dataset into feature and label
y= df['quality']
x=df.drop(columns = ['quality'])
```

```python
x
```

so now removing outliers using zscore and iqr techniques

```python
from scipy.stats import zscore
out_features= x
z=np.abs(zscore(x))
z
```

```python
#treshold=3
np.where(z>3)
```

```python
#now removing the data above zscore and creating new dataframe
df1= df[(z<3).all(axis=1)]

df1.shape
```

```python
print("Data loss percentage- ",((df.shape[0]-df1.shape[0])/df.shape[0])*100)
```

```python
df1
```

```python
#now using IQR method

#1st quantile
Q1 = x.quantile(0.25)

#3rd quantile
Q3 = x.quantile(0.75)

#IQR
IQR = Q3 - Q1

df2= df[~((df <(Q1 - 1.5 * IQR))|(df> (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
df2.shape
```

```python
print("Data loss percentage- ",((df.shape[0]-df2.shape[0])/df.shape[0])*100)
```

As we see that the zscore method will give the less amount of loss percentage data so we used df1 as our new dataframe

# Checking Variance Inflation Factor(VIF)

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif= pd.DataFrame()
vif["VIF values"]= [variance_inflation_factor(x.values, i) for  i in range(len(x.columns))]
vif["Features"]= x.columns
vif
```

Here we can take Threshold Value as '5'

by checking VIF values we can find the feature causing the multicollinarity problem.
here we can find that fixed acidity and density haveVIF value greater than 5.so we will drop one of the column first,if the same problem persist  then we will trying to remove other higher vif value column.

```python
#dropping fixed acidity column
x.drop("fixed acidity", axis=1, inplace=True)
```

```python
#again checking vif values
vif= pd.DataFrame()
vif["VIF values"]= [variance_inflation_factor(x.values, i) for  i in range(len(x.columns))]
vif["Features"]= x.columns
vif
```

# Modelling

Finding the best random State

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
maxAccu = 0
maxRs=0
for i in range(1,200):
    x_train, x_test, y_train,y_test = train_test_split(x1,y1,test_size=0.30, random_state = i)
    RFR = RandomForestClassifier()
    RFR.fit(x_train, y_train)
    pred = RFR.predict(x_test)
    acc = accuracy_score(y_test,pred)
    if acc>maxAccu:
        maxAccu = acc
        maxRs= i
print("Best accuracy is", maxAccu, "at random_state", maxRs)
```

The best accuracy score is 99.78% at the Random State 41

Creating train test split

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.30, random_state= maxRs)
```

# Classification algorithemes

```python
from sklearn.ensemble import RandomForestClassifier,ExtraTreesClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier, AdaBoostClassifier, BaggingClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, accuracy_score
from sklearn.model_selection import cross_val_score
```

RandomForestClassifier

```python
#checking accuracy for RandomForestClassifier
RFC = RandomForestClassifier()
RFC.fit(x_train,y_train)
predRFC = RFC.predict(x_test)
print(accuracy_score(y_test, predRFC))
print(confusion_matrix(y_test,predRFC))
print(classification_report(y_test,predRFC))
```
